```python
import pandas as pd
import joblib
```

```python
# Load the file
loaded_objects = joblib.load("model_and_data.pkl")

# Access items
loaded_model = loaded_objects["model"]
loaded_scaled_data = loaded_objects["scaled_data"]
loaded_input = loaded_objects["input_file"]

print("Loaded model:", loaded_model)
print("Loaded scaled data shape:", loaded_scaled_data)
```

```python
inp = loaded_input
inp.head(10)
```

```python
loaded_scaled_data
```

```python
def ShowCustomerdata():
    print("Model Built on Below Input Data:")
    display(inp.head())

    print("\n--- Pleas Enter Details ---")
    Income = int(input("Enter your incom: "))
    MntWines = int(input("Pleas Enter MntWines products : "))
    MntFruits = int(input("Pleas Enter MntFruits products: "))
    MntMeatProducts = int(input("Pleas Enter MntMeat Products: "))
    MntFishProducts = int(input("Pleas Enter MntFish Products: "))
    MntSweetProducts = int(input("Pleas Enter MntSweet Products: "))
    MntGoldProds = int(input("Pleas Enter MntGold Products : "))
    NumDealsPurchases = int(input("Pleas Enter Number of  Deals Purchases: "))
    NumWebPurchases = int(input("Pleas Enter Number of  Web Purchases: "))
    NumCatalogPurchases = int(input("Pleas Enter Number of  Catalog Purchases: "))
    NumStorePurchases = int(input("Pleas Enter Number of Store Purchases : "))
    NumWebVisitsMonth = int(input("Pleas Enter Number of  WebVisits: "))
    Total_purchas = NumDealsPurchases + NumWebPurchases + NumCatalogPurchases + NumStorePurchases + NumWebVisitsMonth
    Total_spend_product = MntWines + MntFruits + MntMeatProducts + MntFishProducts + MntSweetProducts + MntGoldProds
    display("Total purchas: ",Total_purchas)
    display("Total spend product: " ,Total_spend_product)

    row = pd.DataFrame([[Income,MntWines,MntFruits,
                        MntMeatProducts,
                        MntFishProducts,MntSweetProducts,MntGoldProds,
                        NumDealsPurchases,NumWebPurchases,NumCatalogPurchases,
                        NumStorePurchases,NumWebVisitsMonth,Total_purchas,Total_spend_product]], columns=inp.columns)
    print("Given Data:")
    display(row)

    row = loaded_scaled_data.transform(row)
    cluster_num = loaded_model.predict(row)[0]
   
    print("clustering group number :",cluster_num )
    cluster_labels = {
    0: "Medium income & medium spending customer",
    1: "Low income & low spending customer",
    2: "High income & High spending customer"
    }
    print("Label:", cluster_labels.get(cluster_num, "Unknown cluster"))
        
```

```python
ShowCustomerdata()
```
